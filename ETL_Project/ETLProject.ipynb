```python
#lets explore data

import pandas as pd
orders_df=pd.read_csv('orders.csv')
orders_df
```

```python
#lets check for data quality issues

orders_df.isna().sum()
```

```python
orders_df.columns
# column names have space in between them which is not a good practice. Lets us replace space with '_' using a lambda function

orders_df.columns = orders_df.columns.map(lambda x: x.lower().replace(" ","_"))
orders_df.columns
```

```python
#is the data in a standard format?

#lets explore categorical values.

# unique_val=pd.unique(orders_df[['Ship Mode','Segment','City','State','Category','Sub Category']].values.ravel('K'))
print(orders_df['ship_mode'].unique())
print(orders_df['segment'].unique())
print(orders_df['state'].unique())
print(orders_df['category'].unique())
print(orders_df['sub_category'].unique())

```

```python
#We see that the values in Ship Mode are not standardized. They have values like Not available, unknown.
#let us standardize these values

orders_df['ship_mode'].unique()
```

```python
#lets replace 'Not available' and 'unkown' to 'Not available'

orders_df['ship_mode']= orders_df['ship_mode'].replace(['Not Available','unknown'],'Not Available')
orders_df['ship_mode'].unique()
```

```python
# lets fetch the record for which shiop mode is missing

orders_df[orders_df['ship_mode'].isna()]
```

```python
#steps I took to fill missing info
# Checked what other records in Tennesse, Bristol have as the ship mode for corporate segment
# other records had "standard class" as the shipment record. So filled the N/A with "Standard class" 

fill_value='Standard class'
orders_df['ship_mode']=orders_df['ship_mode'].fillna(fill_value)

orders_df[orders_df['ship_mode'].isna()]

#Missing values have been handled

```

```python
#Lets derive columns like discount, sale price and profit.

#1. Discount 

orders_df['discount']=orders_df['list_price']*(orders_df['discount_percent']/100) # or multiply by 0.01
orders_df.head()
```

```python
#2. Sale price

orders_df['sale_price']=orders_df['list_price']-orders_df['discount']
orders_df.head()
```

```python
#3. Profit

orders_df['profit']=orders_df['sale_price']-orders_df['cost_price']
orders_df.head()
```

```python
# we have to convert datatype of date to datetime. Currently it is in string format

#pd.to_datetime(orders_df['order_date'], format="%Y-%m-%d")

#got an error because the date column in excel sheet is in mm-dd-yy format.
import datetime
#orders_df['order_date'].dt.strftime('%Y%m%d')



orders_df['order_date'] = pd.to_datetime(orders_df['order_date'], format='%Y-%m-%d')
orders_df.head()
orders_df.dtypes
```
